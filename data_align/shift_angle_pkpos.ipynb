```python
# Define the trapezoid window. w is the width of the raising/lower part 
def trape_window(lis, start, end, w):
    window = []
    for x in lis:
        coor = np.array([x])
        if x < start or x > end:
            coor = np.append(coor, [0])
        elif x >= start and x < start+w:
            coor = np.append(coor, [(x - start)/w])
        elif x >= start+w and x <= end-w:
            coor = np.append(coor, [1])
        elif x > end-w and x <= end:
            coor = np.append(coor, [(end - x)/(w)])
        else:
            coor = np.append(coor, [0])
        window.append(coor)
    return np.array(window)

# Plot the window
x = np.linspace(0, 50, 51)
window = trape_window(x, 10,40,5)
plt.figure("taper window")
plt.plot(window[:,0], window[:,1], label='Windowed Signal')
plt.axhline(y=0, color='r', linestyle='--')
plt.legend()
plt.show()
```

```python
# Define a function that minimize the difference between two spectra
def shift_energy(E_ref, ref, E_data, data, lower_lim, upper_lim, plot=False):
    # Change everything into angle space
    angle_ref = f.convert_to_angle(E_ref, 0.0653269)
    angle_data = f.convert_to_angle(E_data, 0.0653269)
    lim = f.convert_to_angle([lower_lim, upper_lim], 0.0653269)
    
    # Find the range of index of angle_ref where we want to integrate over with
    min_ind = np.argmin((abs(angle_ref-lim[0])))
    max_ind = np.argmin((abs(angle_ref-lim[1])))
    
    # Define trepozoid window anf apply the window to ref data
    window = trape_window(np.flip(angle_ref), angle_ref[max_ind],angle_ref[min_ind],0.05)
    ref = np.flip(np.flip(ref) * window[:,1])
    
    # Calculate the correlation function
    correlation = []
    for ang in np.linspace(-0.5,0.5,100):
        spline = CubicSpline(np.flip(angle_data+ang), np.flip(data))
        interp_data = spline(np.flip(angle_ref))
        interp_data = interp_data * window[:,1]
        interp_data = np.flip(interp_data)
        corr = sum(ref[min_ind:max_ind]*interp_data[min_ind:max_ind])
        correlation.append(np.array([ang, corr]))
    correlation = np.array(correlation)
    shift = correlation[:,0][np.argmax(correlation[:,1])]
    
    if plot==True:
        plt.figure("Correlation")
        plt.xlabel('angle shift')
        plt.ylabel('Corr')
        plt.plot(correlation[:,0], correlation[:,1])
        plt.plot([shift, shift], [np.min(correlation[:,1]), np.max(correlation[:,1])], 'k--')

        print("Angle shift is: " + str(shift))
    
    angle_data = angle_data + shift
    shifted_E_data = f.convert_to_energy(angle_data, 0.0653269)
    
    # map window back to energy space
    window_x = np.flip(f.convert_to_energy(window[:,0], 0.0653269))
    window_y = np.flip(window[:,1])
    
    return shifted_E_data, window_x, window_y
```

```python
# Creat a data frame with neighbouring averaged smoothed data wrt to shifted energy
def smooth(data, n):
    data = np.array(data)
    k = np.empty(n*2+1)
    k.fill(1)
    k = k/(n*2+1)
    # getting out a list with dimension len(data)-2n (i.e not including endpoints)
    output = np.convolve(data, k, mode='valid')
    
    # add in end points (which will only average amoung neigbouring points which are in the range)
    for x in range(0,n):
        # left endpoints
        l_list = data[:n+x+1]
        l_ave = sum(l_list)/len(l_list)
        output = np.insert(output, x, l_ave)
        # right endpoints
        r_list = data[-(n+1+x):]
        r_ave = sum(r_list)/len(r_list)
        output = np.insert(output, len(output)-x, r_ave)
    return output

# Put the smoothed, shifted spectra into a sperate data frame
df_smoothed_shifted = pd.DataFrame()
#Set number of neighbouring point on each side to average
n = 1
for file in file_list:
    df_smoothed_shifted['Energy:' + file] = df['shifted E:'+file]
    df_smoothed_shifted[file] = smooth(df[file], n)
    
df_smoothed_shifted
    
```
