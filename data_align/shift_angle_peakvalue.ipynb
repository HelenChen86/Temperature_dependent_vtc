## This program shifts data given the k-beta peak positions before and after shift

```python
import numpy as np
import matplotlib.pyplot as plt
import functions_peakvalue as fun
#%matplotlib notebook
```

Read in data

```python
energy, data = fun.read_file()

#convert the energy values to angle
angle = fun.convert_to_angle(energy)  # d for Ge 555 is 0.0653269 nm

#convert the lists to arrays  
angle = np.array(angle, dtype=float)
data = np.array(data, dtype=float)
```

Shift in angle and convert to energy

```python
shift = fun.convert_shift(9572.00, 9576.25)

angle_shifted = fun.spectra_shift(angle, shift)

energy_shifted = fun.convert_to_energy(angle_shifted)
```

```python
fun.write_to_file(energy_shifted, data, 'file_ref', delimiter='\t')
```

```python
plt.figure(figsize=(6,4))
plt.plot(energy, data, energy_shifted, data)
plt.legend(['data', 'shifted_data'], loc = 'best')
plt.axis([9560, 9590, 0, 0.12])
```
